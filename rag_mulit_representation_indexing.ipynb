```python
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
from langchain.prompts import ChatPromptTemplate
prompt=ChatPromptTemplate.from_template("Summarize the data of docs:{docs}")
llm=ChatGroq(temperature=0)

chain=(
    {"docs": lambda x: x.page_content}
    |prompt
    |llm
    |StrOutputParser()

)
```

```python
summaries = chain.batch(docs, {"max_concurrency": 1})
```

```python
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.embeddings import HuggingFaceBgeEmbeddings
from langchain.vectorstores import Chroma
import uuid

from langchain.storage import InMemoryByteStore
from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-small-en"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
hf_embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)

```

```python
!pip -q install  chromadb
```

```python
vectorstore=Chroma(
    embedding_function=hf_embeddings,
    collection_name="summaries"
)
```

```python
# the docstore or bytestore
from langchain.storage import InMemoryByteStore
docstore=InMemoryByteStore()
id_key="doc_id"
```

```python
# The retriever
retriever=MultiVectorRetriever(
    vectorstore=vectorstore,
    byte_store=docstore,
    id_key=id_key
)
```

```python
#generate ids for each doc in docs
doc_ids=[str(uuid.uuid4()) for _ in docs]
```

```python
for i,data in enumerate(summaries):
  print(f"{i}: : {data}")
```

```python
#docs linking to summaries
from langchain_core.documents import Document
summaries_docs=[

    Document(page_content=data,metadata={id_key:doc_ids[i]})
    for i,data in enumerate(summaries)
]
```

```python
summaries_docs
```

```python
# Adding
retriever.vectorstore.add_documents(summaries_docs)
retriever.docstore.mset(list(zip(doc_ids, docs)))
```

```python
query = "What is llm allignment?"
sub_docs = vectorstore.similarity_search(query,k=1)
sub_docs[0].page_content
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
rag_temp="""
You are an Ai assistant ,You generate answer to the input query based on the context.
here is the context:{context}
here is the query:{query}
answer:
"""
```

```python
rag_prompt=ChatPromptTemplate.from_template(rag_temp)
```

```python
search_chain = (
    {"query": RunnablePassthrough(), "context": lambda x: vectorstore.similarity_search(x, k=1)}  # Access query directly
    | rag_prompt
    | llm
    | StrOutputParser()
)
```

```python
query = "What is llm allignment?"
search_chain.invoke(query)
```

```python
def search(query):
  return search_chain.invoke(query)
```

```python
query="what is generative ai?"
search(query)
```
